# Octave Tutorial

### Elementary operations:

* __Arithmetic__ __operations__: Same as in Python, except ^ is exponentiation.  
* __Comparision__ __operations__: Same as in Python, except ~ is not.  
* __Logical__ __operations__: Same as in Java.  
* __Bitwise__ __operations__: __xor(a, b)__ function is xor.

### Manipulating data:

```octave
A = [1, 2; 3 4; 5 6]

a32 = A(3, 2) % getting an element

a_row_3 = A(3, :) % getting a row

a_col_2 = A(:, 2) % getting a column

A([1 3], :) % get all elements on row 1 and 3

A(:, 1) = [1;2;3] % change first column

A = [A, [4; 5; 6]] % append a column to the right

A(:) % creates a vector of all the elements in A

B = [24 12; 64 2; 33 6];
C = [A B] % concatenate A and B into a new matrix
C = [A(:, 1:2); B]
```

### Computing on data:

```octave
A = [1 2; 3 4; 5 6];
B = [11 12; 13 14; 15 16];
C = [1 1; 2 2];
```

```octave
A * C % matrix multiplication
A .* B % element multiplication
A .^ 2 % element wise squaring of A
```

```octave
v = [1; 2; 3];

1 ./ v % element wise reciprocal of v
log(v)
exp(v)
abs(v)
-v % -1 * v
v + ones(3, 1)
v' % transpose of v
max(v) % maximum element on each column
[val, index] = min(v)
v < 1 % element wise comparison
find(v > 1) % returns the elements bigger that 1
sum(v)
prod(v)
floor(v)
ceil(v)

magic(3) % returns a 3 x 3 magic square
```

### Plotting data:

```octave
t = [0:0.01:0.98]; % row vector

y1 = sin(2 * pi * 4 * t);
y2 = cos(2 * pi * 4 * t);

plot(t, y1); % t - x axis; y1 - y axis
hold on; % will plot both graphs on the same figure
plot(t, y2, 'r'); % red color
xlabel('time');
ylabel('value');
legend('sin', 'cos');
title('very interesting plot :D');
% print -dpng 'my_plot.png' % to save the figure
```

```octave
% displaying to graphs in different figures
figure(1); plot(t, y1);
figure(2); plot(t, y2);
```

```octave
% displaying to graphs in the same window
subplot(1, 2, 1); % divides the plot into 1 x 2 grid, and accesses the first element
plot(t, y1);
subplot(1, 2, 2);
plot(t, y2);
axis([0.5 1 -2 2]) % changes x axis range to 0.5 to 1, and y axis to -2 to 2
% clf; % clear figure
```

### Control Statements:

```octave
v = zeros(10, 1);

for i = 1:10,
    v(i) = 2 ^ i;
end;

for i=v,
    disp(i);
end;

% break and continue are allowed inside for statements
```

```octave
i = 1;
while i <= 5,
    v(i) = 100;
    i = i + 1;
end;
v
```

```octave
i = 1;
while true,
    
    if i == 6,
        break;
    elseif i == 3,
        v(i) = v(i) ^ 2;
    else
        v(i) = 999;
    end;
    
    i = i + 1;
end;
v
```

```octave
function y = squareIt (x) % y is the return variable
    y = x ^ 2;
endfunction;
```

```octave
squareIt(5)
```

```octave
function [y, z] = squareAndCube(x)
    y = x ^ 2;
    z = x ^ 3;
endfunction;
```

```octave
[a, b] = squareAndCube(3)
```

```octave
f = @(x) x .^ 2;
plot(1:0.1:10, f(1:0.1:10))
```
